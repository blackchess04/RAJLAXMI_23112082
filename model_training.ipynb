```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

```

```python
X = pd.read_csv("X_train.csv")
y = pd.read_csv("y_train.csv").values.ravel()
X_test = pd.read_csv("X_test.csv")

```

```python
X
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X.drop(columns=["id"]),
    y,
    test_size=0.2,
    random_state=42
)

```

```python
X_train
```

```python
y_train
```

```python
model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
model.fit(X_train, y_train)
```

```python
val_preds = model.predict(X_val)

r2 = r2_score(y_val, val_preds)
rmse = np.sqrt(mean_squared_error(y_val, val_preds))

print("R2 Score:", r2)
print("RMSE:", rmse)
```

```python

rmse = np.sqrt(mean_squared_error(y_val, val_preds))
mean_price = y_val.mean()

print("RMSE %:", rmse / mean_price * 100)

```

```python
test_preds = model.predict(X_test.drop(columns=["id"]))

final_df = pd.DataFrame({
    "id": X_test["id"],
    "predicted_price": test_preds.astype(int)
})

final_df.to_csv("enrollno_final.csv", index=False)
final_df.head()
```
